Find the percentage of games that went to extra time. 

```python
df[df['Score'].apply(lambda x: 'a.e.t' in x)]
```

```python
len(df[df['Score'].apply(lambda x: 'a.e.t' in x)])
```

```python
# To find the percentage of matches that had extrra time.

np.round((len(df[df['Score'].apply(lambda x: 'a.e.t' in x)]) / len(df) * 100),2)
```

```python
# How many go to penalties
len(df[df['Score'].apply(lambda x: 'pen.' in x)])
```

```python
np.round((len(df[df['Score'].apply(lambda x: 'pen' in x)]) / len(df[df['Score'].apply(lambda x: 'a.e.t' in x)]) * 100),2)
```
